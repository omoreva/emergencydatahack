# Classification task using logistic regression

* Logistic regression could be used for binary classification
* For multiclass classification sklearn uses one-vs-rest (OvR) method
* Good for the linearly separable classes

As we see from plots, iris data is linearly separated, so logistic regression algorithm will work great here.

```python
# prepare X, y for model training

X = df.copy()
del X['label']
X = X.to_numpy()


y = iris_dataset['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lr = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
y_pred = lr.predict(X_test) 
```

```python
print(classification_report(y_test, y_pred, target_names=iris_dataset['target_names']))
```

The accuracy of the model on the test data is great, precision, recall and f1-score are perfect. Let's try to use another dataset to see how logistic regression will deal with more complicated data

```python
X = wine_dataset['data']
y = wine_dataset['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=48)
```

```python
# prepare configuration for cross validation test harness
seed = 7
# prepare models
models = []
models.append(('SupportVectorClassifier', SVC()))
models.append(('StochasticGradientDecentC', SGDClassifier()))
models.append(('RandomForestClassifier', RandomForestClassifier()))
models.append(('DecisionTreeClassifier', DecisionTreeClassifier()))
models.append(('GaussianNB', GaussianNB()))
models.append(('KNeighborsClassifier', KNeighborsClassifier()))
models.append(('AdaBoostClassifier', AdaBoostClassifier()))
models.append(('LogisticRegression', LogisticRegression()))
# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
   kfold = KFold(n_splits=10, random_state=seed)
   cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
   results.append(cv_results)
   names.append(name)
   msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
   print(msg)
```

```python
# boxplot algorithm comparison
fig = plt.figure(figsize=(20, 12))
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

The best model appeared to be GaussianNB
